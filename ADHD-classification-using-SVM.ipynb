## Dimensionality reduction using PCA

```python
class PCA:
    def __init__(self, n_components):
        self.n_components = n_components
    
    def fit(self, X):
        cov = np.cov(X.T)

        e_vals, e_vecs = np.linalg.eig(cov)
        p = np.argsort(e_vals)[::-1]
        e_vecs = e_vecs[p]
        
        self.components_ = e_vecs[:self.n_components, :]

    def transform(self, X):
        if(self.components_ is not None):
            mean_sub = X - np.mean(X, axis=0)
            return(mean_sub @ self.components_.T)
```

```python
convert = PCA(2)
convert.fit(data)
converted_data = convert.transform(data)
```

```python
converted_data['Class'] = [0,0,1,1]
converted_data
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(converted_data.drop(['Class'],axis =1), converted_data['Class'])
```

```python
from sklearn.metrics import accuracy_score
predicted = clf.predict(converted_data.drop(['Class'],axis =1))
print(accuracy_score(predicted,converted_data['Class']))
```
